```python
def matrices_to_cov(matrix_dictionary):
    headers = ['chromosome','coordinate', 'methylation percentage', 'count methylated', 'count unmethylated', 'total']
    chromosomes = []
    coordinates = []
    methylation_percentages = []
    counts_methylated = []
    counts_unmethylated = []
    totals = []

    for key, value in matrix_dictionary.items():
        cpg_coords = list(value.columns)
        transpose = value.to_numpy().transpose() 
        for i in range(len(cpg_coords)):
            
            chromosomes.append(key.split('_')[0])
            coordinates.append(cpg_coords[i] + 1)
            num_unmethylated = (transpose[i] == 0).sum()
            num_methylated = (transpose[i] == 1).sum()
            methylation_percentages.append(100.0 * num_methylated/(num_methylated + num_unmethylated))
            counts_methylated.append(num_methylated)
            counts_unmethylated.append(num_unmethylated)
            totals.append(num_methylated + num_unmethylated)
        
    
    list_of_tuples = list(zip(chromosomes, coordinates, methylation_percentages, counts_methylated, counts_unmethylated, totals))
    df = pd.DataFrame(list_of_tuples, columns=headers)
    
    return df
```

**NOTE : Please note that the column "total" means count methylated + count unmethylated.**

```python
wbc3842_chr5_pre_imputation_df = matrices_to_cov(wbc3842_chr5_pre_impute_matrices)
wbc3842_chr5_pre_imputation_df[0:50]
```

```python
wbc3842_chr5_imputed_cpg2_df = matrices_to_cov(wbc3842_chr5_imputed_cpg2)
wbc3842_chr5_imputed_cpg3_df = matrices_to_cov(wbc3842_chr5_imputed_cpg3)
wbc3842_chr5_imputed_cpg4_df = matrices_to_cov(wbc3842_chr5_imputed_cpg4)
wbc3842_chr5_imputed_cpg5_df = matrices_to_cov(wbc3842_chr5_imputed_cpg5)

```

```python
# making cpg_index as the index of the df
wbc3842_chr5_pre_imputation_df = wbc3842_chr5_pre_imputation_df.reset_index(drop=True).set_index('coordinate', drop=False)

# making cpg_index as the index of the df as imputed ones are not sorted
wbc3842_chr5_imputed_cpg2_df = wbc3842_chr5_imputed_cpg2_df.reset_index(drop=True).set_index('coordinate', drop=False)
wbc3842_chr5_imputed_cpg2_df.sort_index(inplace=True)

# making cpg_index as the index of the df as imputed ones are not sorted
wbc3842_chr5_imputed_cpg3_df = wbc3842_chr5_imputed_cpg3_df.reset_index(drop=True).set_index('coordinate', drop=False)
wbc3842_chr5_imputed_cpg3_df.sort_index(inplace=True)

# making cpg_index as the index of the df as imputed ones are not sorted
wbc3842_chr5_imputed_cpg4_df = wbc3842_chr5_imputed_cpg4_df.reset_index(drop=True).set_index('coordinate', drop=False)
wbc3842_chr5_imputed_cpg4_df.sort_index(inplace=True)

# making cpg_index as the index of the df as imputed ones are not sorted
wbc3842_chr5_imputed_cpg5_df = wbc3842_chr5_imputed_cpg5_df.reset_index(drop=True).set_index('coordinate', drop=False)
wbc3842_chr5_imputed_cpg5_df.sort_index(inplace=True)
```

```python
# replacing pre-impute rows with imputed ones if available

wbc3842_chr5_pre_imputation_df.loc[wbc3842_chr5_pre_imputation_df['coordinate'].isin(wbc3842_chr5_imputed_cpg2_df['coordinate']), ['methylation percentage', 'count methylated','count unmethylated', 'total']] = wbc3842_chr5_imputed_cpg2_df[['methylation percentage', 'count methylated','count unmethylated', 'total']]
wbc3842_chr5_pre_imputation_df.loc[wbc3842_chr5_pre_imputation_df['coordinate'].isin(wbc3842_chr5_imputed_cpg3_df['coordinate']), ['methylation percentage', 'count methylated','count unmethylated', 'total']] = wbc3842_chr5_imputed_cpg3_df[['methylation percentage', 'count methylated','count unmethylated', 'total']]
wbc3842_chr5_pre_imputation_df.loc[wbc3842_chr5_pre_imputation_df['coordinate'].isin(wbc3842_chr5_imputed_cpg4_df['coordinate']), ['methylation percentage', 'count methylated','count unmethylated', 'total']] = wbc3842_chr5_imputed_cpg4_df[['methylation percentage', 'count methylated','count unmethylated', 'total']]
wbc3842_chr5_pre_imputation_df.loc[wbc3842_chr5_pre_imputation_df['coordinate'].isin(wbc3842_chr5_imputed_cpg5_df['coordinate']), ['methylation percentage', 'count methylated','count unmethylated', 'total']] = wbc3842_chr5_imputed_cpg5_df[['methylation percentage', 'count methylated','count unmethylated', 'total']]


wbc3842_chr5_replaced = wbc3842_chr5_pre_imputation_df
# df with cpgs having imputed reads replaced 
print(wbc3842_chr5_replaced.isna().sum())
wbc3842_chr5_replaced.head(30)
```
